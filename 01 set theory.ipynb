3: Set operation

Set Operation 1:

C⊕DC⊕D

The symmetric difference (⊕⊕) of two sets is the set of elements that are in either of the sets but not in their intersection.

C⊕D=(C\D)∪(D\C)C⊕D=(C\D)∪(D\C)

C\D={6,7,8,9}C\D={6,7,8,9}

D\C={1,3}D\C={1,3}

So, C⊕D={1,3,6,7,8,9}C⊕D={1,3,6,7,8,9}

# Define the sets

D = {1, 3, 5, 7, 9}
